```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link = 'https://www.amazon.in/product-reviews/B06ZZ4G9ZM'
```

```python
page = requests.get(link)
page
```

```python
page.content
```

```python
soup = bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span', class_='a-profile-name')
names
```

```python
names.pop(0)
```

```python
names.pop(0)
```

```python
names
```

```python
 len(names)
```

```python
cust_name = []
for i in range (0, len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
title = soup.find_all('a', class_ = 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
title
```

```python
review_title = []
for i in range (0, len(title)):
    review_title.append(title[i].get_text())
```

```python
review_title
```

```python
review_title[:] = [title.lstrip('\n') for title in review_title] 
review_title[:] = [title.rstrip('\n') for title in review_title]
review_title
```

```python
rating = soup.find_all('i', class_='review-rating')
## rating = soup.find_all('i', {"data-hook":"review-rating"})
rating
```

```python
review_rating = []
for i in range (0, len(rating)):
    review_rating.append(rating[i].get_text())
```

```python
review_rating
```

```python
review_rating.pop(0)
review_rating.pop(0)
```

```python
review_rating
```

```python
content = soup.find_all('span', class_ = 'a-size-base review-text review-text-content')
## content = soup.find_all('i', {"data-hook":"review-body"})
content
```

```python
review_content = []
for i in range (0, len(content)):
    review_content.append(content[i].get_text())
```

```python
review_content
```

```python
review_content[:] = [content.lstrip('\n\n') for content in review_content] 
review_content[:] = [content.rstrip('\n\n') for content in review_content]
review_content
```

```python
len(review_content)
```

```python
df = pd.DataFrame()
```

```python
df['Customer Name'] = cust_name
df['Review Title'] = review_title
df['Review Rating'] = review_rating
df['Review Content'] = review_content
```

```python
df
```
